### PCA

```python
import numpy as np
np.linalg.eig(corr)
```

```python
# scree plot
# eigen values가 1보다 낮은건 버리기
from matplotlib import pyplot as plt
eigen_val = np.linalg.eig(corr)[0]
np.sort(eigen_val)
index=['PC1', 'PC2', 'PC3', 'PC4', 'PC5','PC6', 'PC7', 'PC8', 'PC9']
plt.plot(index, eigen_val, 'ro-')
plt.xlabel('Component Number')
plt.ylabel('Variances')
plt.title('Covariance Scree Diagram')
plt.axhline(y=1, color='b', linewidth=2)
plt.grid(True)
plt.show()
```

### MLE

```python
from factor_analyzer import FactorAnalyzer
fa = FactorAnalyzer(n_factors=2, rotation=None, method='ml')  # method : MLE
fa.fit(data1)
efa = pd.DataFrame(fa.loadings_, index=data1.columns)  # loadings
efa
```

```python
import seaborn as sns
sns.heatmap(efa, cmap='Blues', annot=True, fmt='.2f')  # lodings 
```

```python
from factor_analyzer import FactorAnalyzer
fa = FactorAnalyzer(n_factors=2, rotation='varimax', method='ml')  # method : MLE
fa.fit(data1)
efa1 = pd.DataFrame(fa.loadings_, index=data1.columns)  # loadings
efa1
```

```python
import seaborn as sns
sns.heatmap(efa1, cmap='Blues', annot=True, fmt='.2f')  # lodings 
```

### principal factor with iterative

```python
from factor_analyzer import FactorAnalyzer
fa = FactorAnalyzer(n_factors=2, rotation='varimax', method='principal')  # method : MLE
fa.fit(data1)
efa2 = pd.DataFrame(fa.loadings_, index=data1.columns)  # loadings
efa2
```

```python
import seaborn as sns
sns.heatmap(efa2, cmap='Blues', annot=True, fmt='.2f')  # lodings 
```
